```python
import cv2
import numpy as np
# Import cv2_imshow from google.colab.patches
from google.colab.patches import cv2_imshow

img = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.putText(img, 'Test Window', (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
# Use cv2_imshow instead of cv2.imshow
cv2_imshow(img)
# cv2.waitKey(0) and cv2.destroyAllWindows() are not needed with cv2_imshow
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# ✅ Step 1: Upload an image (if you don't already have one)
from google.colab import files
uploaded = files.upload()  # Choose an image like 'example.jpg'

# ✅ Step 2: Load the image using OpenCV
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# ✅ Step 3A: Display using OpenCV's cv2_imshow (for BGR image)
print("Image shown with cv2_imshow:")
cv2_imshow(image)

# ✅ Step 3B: Display using matplotlib (requires conversion BGR to RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6, 6))
plt.imshow(image_rgb)
plt.title("Image Displayed with matplotlib")
plt.axis('off')
plt.show()
```
